# Learn NUMPY in 5 minutes - BEST Python Library!

```python
import matplotlib.pyplot as plt
from skimage import io
photo = io.imread('100x160.jpg')

```

```python
photo.shape
```

```python
plt.imshow(photo)
```

```python
plt.imshow(photo[::-1])
```

```python
plt.imshow(photo[:, ::-1])
```

```python
plt.imshow(photo[20:100, 25:80])
#[row:row, col:col]
```
