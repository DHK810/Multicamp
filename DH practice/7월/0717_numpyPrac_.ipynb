# Learn NUMPY in 5 minutes - BEST Python Library!

```python
import matplotlib.pyplot as plt
from skimage import io
photo = io.imread('100x160.jpg')

```

```python
photo.shape
```

```python
plt.imshow(photo)
```

```python
plt.imshow(photo[::-1])
```

```python
plt.imshow(photo[:, ::-1])
```

```python
plt.imshow(photo[20:100, 25:80])
#[row:row, col:col]
```

```python
import numpy as np
photo_sin = np.sin(photo)
photo_sin
```

```python
print(np.sum(photo))
print(np.prod(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.var(photo))
```

```python
photo_masked = np.where(photo >100, 255, 0)
```

```python
plt.imshow(photo_masked)
```

```python
plt.imshow(photo[:,:,0].T)
```
