```python
import requests
from bs4 import BeautifulSoup

url = 'https://search.naver.com/search.naver?where=post&sm=tab_jum&query=%EB%A8%B8%EC%8B%A0%EB%9F%AC%EB%8B%9D'
resp = requests.get(url)

soup = BeautifulSoup(resp.content, 'html.parser')

ul_contents = soup.find('div', class_ = 'blog section _blogBase _prs_blg')
ul_contents
ul_contents.contents  #리스트 첫번째[0]에 공백이 있음


new_ul = []
for content in ul_contents.contents:
    if not str(content).strip():  #공백 제거 한것이 None이면 continue
        continue
    new_ul.append(content)

print(new_ul)
new_list = []
for li in new_ul:
    title = li.find('ul').find('li').find('dl').find('dt')
    new_list.append(title)

print(new_list)
```

# 성공

```python

from bs4 import BeautifulSoup

url = 'https://search.naver.com/search.naver?where=post&sm=tab_jum&query=%EB%A8%B8%EC%8B%A0%EB%9F%AC%EB%8B%9D'
resp = requests.get(url)

soup = BeautifulSoup(resp.content, 'html.parser')

ul_contents = soup.find('ul', class_ = 'type01') #크롤링 하고자 하는 부분의 최고 프레임을 설정
li_tag = ul_contents.find_all('li')         #10개의 블로그 제목들을 포함하는 태그 가져오기
print(li_tag)

news = []
for li in li_tag:           
    new_lst = {}       
    a_tag = li.find('dl').find_all('dd')   #url을 가진 dd태그에 접근
    new_lst['title'] = li.find('dt').text  #제목을 가진 dt태그에 접근
    new_lst['summary'] = a_tag[1].text     #요약을 가진 a태그에 접근
    news.append(new_lst)
    
news
```
