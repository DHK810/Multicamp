# 블로그 내용 가져오기

```python
blog_url = 'https://search.naver.com/search.naver'

params = dict(
    where = 'post',
    query = '놀란 라이언'
    )

```

```python
resp = requests.get(blog_url, params = params)
resp
```

```python
soup = BeautifulSoup(resp.content)
blog_list = soup.find_all('li', class_ ='sh_blog_top') #해당 클래스를 가진 li 태그들을 다 보여줘
len(blog_list)

for blog in blog_list:
    a_tag = blog.find('a')
    print(a_tag)
    link = a_tag['href']
    print(link)
    detail_resp = requests.get(link)
    print(detail_resp)
    detail_soup = BeautifulSoup(detail_resp.content)
    blog_post = detail_soup.find('div', class_ = 'se_doc_viewer se_body_wrap se_theme_transparent')
    print(blog_post)         #none 값. 왜? 블로그 브라우저는 frame별로 그려진다. 블로그의 iframe이 문제, src가 원문 주소
    
    break
```
