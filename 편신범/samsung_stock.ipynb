# 네이버 금융에서 주가 여러페이지 가져오기

```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
URL_samsung= 'https://finance.naver.com/item/sise_day.nhn'
com_code = "005930"
page_list = []
j = 0
stock_list = []
for page in range(1,4):
    
    j+=1
    #페이지 스타트
    
    params = {
        'code': com_code,
        'page': j
    }

    resp = requests.get(URL_samsung, params = params)
    print("접속상태", resp.status_code)
    soup = BeautifulSoup(resp.content, 'html.parser')
    ###########################
    sise_list = []
    sise_list =soup.find_all('tr')

    for sise in sise_list:
        price_dict = {}  # dict는 하나만 만들고, index만 바꿔줌
        date_tag= sise.find('span','tah p10 gray03')

        if date_tag is None:
            continue
        price_dict['Date']= date_tag.text

        price_tag = sise.find_all('span','tah p11')
        i=0
        for p in price_tag:

            i +=1
            
            if p.text is None:
                continue
            elif p.text == '0':
                i -=1
                continue
#                 price_dict['Price{}'.format(i)] = price_dict.get('Price{}'.format(i-1))
#             elif p.text == 0:
#                 continue
            else:
                price_dict['Price{}'.format(i)] = p.text
            
        stock_list.append(price_dict)

#     page_list.append(stock_list)
    pprint(stock_list)
```

```python
import pandas as pd
```

```python
## 받은 list를 pandas 이용해서 csv 저장
# column_list = ['날짜','종가','시가','고가','저가','거래량']
df = pd.DataFrame(stock_list)
# df.rename(columns={'Date':'날짜','Price1':'종가','Price2':'시가','Price3':'고가','Price4':'저가','Price5':'거래량'})
```
