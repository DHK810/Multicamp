```python
import glob
import os
import pandas as pd

results = pd.DataFrame([])

for counter, file in enumerate(glob.glob("*.csv")):
    namedf = pd.read_csv(file)
    results = results.append(namedf)

results.to_csv('output.csv')
```

```python
results = results.sort_values(["review_date"], ascending=[False])

results
```
