```python
import requests
from bs4 import BeautifulSoup


def crawl_KW(url, word):
    word_list = []
    
    for i in range(1, 2):
        url_format = url +'/browse/' +  word + '/' + str(i)
        resp = requests.get(url_format)
        soup = BeautifulSoup(resp.content, 'html.parser')

    
        find_frames = soup.find_all('div', class_='css-10cokgy e1x7e0fw0')

        for section in find_frames:

            li_tags = section.find("div", class_ ='css-rats3j e1qo4u831').find('ul', class_= 'css-k9iupi et6tpn80')#, class_ = 'css-k9iupi et6tpn80').find_all('li')
            for li in li_tags:
                word = li.text
                word_list.append(word)

        if soup.find('li', class_ = 'rc-pagination-disabled rc-pagination-next'):
            break


    return word_list
    
```

```python
url = 'https://www.thesaurus.com'
KW_outperform = crawl_KW(url, 'outperform')
KW_choose = crawl_KW(url, 'choose')
KW_prefer = crawl_KW(url, 'prefer')
KW_superior = crawl_KW(url, 'superior')
KW_inferior = crawl_KW(url, 'inferior')
```

```python
KW = KW_outperform+ KW_choose + KW_prefer + KW_superior + KW_inferior
```

```python
print(len(KW))
print(len(set(KW)))
```

```python
KW_set = set(KW)
```
