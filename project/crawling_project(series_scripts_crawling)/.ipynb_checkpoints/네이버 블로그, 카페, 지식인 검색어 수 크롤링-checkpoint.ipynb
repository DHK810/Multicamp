```python
import requests
from bs4 import BeautifulSoup
import pandas as pd




index_keyword = ['헤어디자이너', '헤어디자이너 취미', '헤어디자이너 지망', '헤어디자이너 되려면', '헤어디자이너가 되고 싶어',
         '요리사', '요리사 취미', '요리사 지망', '요리사 되려면', '요리사가 되고 싶어',
         '웹툰작가', '웹툰작가 취미', '웹툰작가 지망', '웹툰작가 되려면', '웹툰작가가 되고 싶어',
         '일러스트레이터', '일러스트레이터 취미', '일러스트레이터 지망', '일러스트레이터 되려면', '일러스트레이터가 되고 싶어',
         '웹소설 작가', '웹소설 작가 취미', '웹소설 작가 지망', '웹소설 작가 되려면', '웹소설 작가가 되고 싶어']

where_list = ['kin', 'post', 'article'] #kin(지식인), post(블로그), ariticle(카페)

df = pd.DataFrame(columns = ['kin', 'post', 'article'], index = index_keyword)
```

```python

def replace_space(keyword):  
    
    keyword = keyword.replace(" ", "%20") #띄어쓰기는 url주소에 쓰이기 위해서 '-'로 변환
    return keyword

def crawl_naver(term, date_from = '20030209', date_to='20190808'):
    for keyword in index_keyword:
        #지식인과 카페는 블로그랑 다른 날짜 구분 형식을 가지고 있음.
        # 지식인 + 카페 url
        for where in ['kin', 'article']:
            params = {
                'where' : where,
                'query' : keyword,
                'date_option' : term,
                'date_from' : date_from,
                'date_to' : date_to
                }

            #url = 'https://search.naver.com/search.naver?'
            url = 'https://search.naver.com/search.naver?where={}&query={}&kin_sort=0&c_id=&c_name=&sm=tab_opt&sec=0&title=0&answer=0&grade=0&choice=0&nso=so%3Add%2Ca%3Aall%2Cp%3A{}m&ie=utf8'.format(where, keyword, term)

            resp = requests.get(url)
            soup = BeautifulSoup(resp.content, 'html.parser')

            page_no = int(soup.find('span', class_ = 'title_num').text.split('/')[1].replace('건','').replace(',',''))

            df.loc[keyword, where] = page_no
        # 블로그 url
        url_post = 'https://search.naver.com/search.naver?where=post&query={0}&st=date&sm=tab_opt&date_from={1}&date_to={2}&date_option={3}'.format(keyword, date_from, date_to, term)
        resp_post = requests.get(url_post)
        soup_post = BeautifulSoup(resp_post.content, 'html.parser')

        page_no = int(soup.find('span', class_ = 'title_num').text.split('/')[1].replace('건','').replace(',',''))

        df.loc[keyword, 'post'] = page_no
        
    return df
```

```python
#term= 6(6개월), 5(3개월), 4(1개월)
```

```python

df = crawl_naver(6)
```

```python
df.loc[:, 'sum'] = df.sum(axis=1)
```

```python
df.rename(columns = {'kin' : '지식인', 'post' : '블로그' , 'article' : '카페', 'sum' : '합'}, inplace = True)
```
