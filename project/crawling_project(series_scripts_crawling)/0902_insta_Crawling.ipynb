```python
import requests
import json
import urllib.request
import datetime as dt
```

```python
def insta(keyword):
    page = "0"
    after = ""
    while True:
        try:
            url = 'https://www.instagram.com/graphql/query/?query_hash=174a5243287c5f3a7de741089750ab3b&variables={"tag_name":"' + keyword + '","first": ' + page + ',"after": "' + after + '"}'  # 요청할 주소
            resp = requests.get(url)
            json_data = json.loads(resp.text)
            page = str(int(page) + 1)
            i = 0
            print(url)
            after = json_data['data']['hashtag']['edge_hashtag_to_media']['page_info']['end_cursor']
            for edge in json_data['data']['hashtag']['edge_hashtag_to_media']['edges'] :
                try:
                    text = edge['node']['edge_media_to_caption']['edges'][0]['node']['text']
                    print(str(i) + '번째 text : ' + text)

                    timestamp = edge['node']['taken_at_timestamp']
                    date = dt.datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S')
                    print(str(i) + '번째 date : ' + date)

                    shortcode = edge['node']['shortcode']
                    each_url = 'https://www.instagram.com/graphql/query/?query_hash=477b65a610463740ccdb83135b2014db&variables={"shortcode":"' + shortcode + '"}'
                    print(str(i) + '번째 each_url : ' + each_url)

                    ###숫자 형식으로 가져옴
                    like_count = edge['node']['edge_liked_by']['count']
                    print(str(i) + '번째 like_count : ' + str(like_count))


                    explain = edge['node']['accessibility_caption']
                    print(str(i) + '번째 explain : ' + explain)

                    each_resp = requests.get(each_url)
                    each_json_data = json.loads(each_resp.text)
                    if each_json_data['data']['shortcode_media']['location']['name']:
                        each_location = each_json_data['data']['shortcode_media']['location']['name']
                        print(str(i) + '번째 each_location : ' + each_location)
                    else:
                        pass

                    

                    print("\n\n\n\n")

                except Exception as e: # 에러 종류
                    print('에러가 발생 했습니다 e', e)
                    break
        except Exception as e1: # 에러 종류
            print('에러가 발생 했습니다 e1', e1)
            break 
```

```python
insta('운동')
```
