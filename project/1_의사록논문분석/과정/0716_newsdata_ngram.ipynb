# MPCK

```python
import pandas as pd
from ekonlpy.sentiment import MPCK
file_df = pd.read_csv('2005-2017_news_data_final.csv')

def text2ngram(text) :
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    return ngrams

```
