# ngram 함수

```python
import pandas as pd
from ekonlpy.sentiment import MPCK
file_df = pd.read_csv('n_grams.csv')

def text2ngram(text) :
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    return ngrams

```

# 뉴스 본문 n gram 반복문

```python
count = 0
list_text = []
for idx, text in enumerate(file_df['text']):
    
    list_text.append(text2ngram(text))

    if idx % 300 == 0:
        print(idx)
```
