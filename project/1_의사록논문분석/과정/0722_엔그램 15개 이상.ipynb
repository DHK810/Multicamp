# 비어있는 ngrams 추가하기

```python
list_ = []
for idx in file_copy[file_copy['ngrams'] == '[]'].index:
    list_.append(idx)
```

```python
from ekonlpy.sentiment import MPCK
def text2ngram(text) :
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    return ngrams

count = 0
list_text = []
for idx in list_:
    text = file_copy.text[idx]
    list_text.append(text2ngram(text))
    count += 1
    if count % 50 == 0:
        print(count)
```
