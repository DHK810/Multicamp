# BYC를 위해 새로운 데이터 프레임 생성
# ngram + up_down만 뽑아냄

```python
ngram_updown = pd.DataFrame(file_copy[['ngrams', 'up_down']][:3200])
```

```python
training_set = np.array(ngram_updown)
```

```python
training_set
```

```python
# 1 = 1
# 0 = 0
# 2 = -1
doccnt1 = 0
doccnt2 = 0
wordfreq = defaultdict(lambda : [0, 0]) 
print(wordfreq)
#defaultdict = lambda - key값, value - (x, y)값으로 반환함.
for ngrams, up_down in training_set:
    #trainint_set = mail + label
    #doc = key = mail, point = span/non-spam = label
    words = ngrams.split(',')
    
    for word in words:
        if up_down == 1: # 상승
            wordfreq[word][0] += 1
            #wordfreq(key)[x값]
            #x값 = 스팸
            #word가 스팸일 경우 wordfreqp[word][x] 증가
        elif up_down == -1: # 하락
            wordfreq[word][1] += 1
            #wordfreq(key)[y값]
            #y값 = 논스팸
            #word가 논스팸일 경우 wordfreqp[word][y] 증가
        
            
print(wordfreq)
# wordfreq 내부 데이터는 key : (x, y)로 되어있음
for key, (cnt1, cnt2) in wordfreq.items():
    
    if cnt1 >0:
        doccnt1 += 1
    #cnt1 = 스팸 중 key가 들어간 빈도
    #doccnt1 = 스팸에 속하는 모든 단어들 수
   
    if cnt2 > 0:
        doccnt2 += 1
    
     #cnt0 = 논스팸
     #doccnt1 = 논스팸에 속하는 모든 단어들 수
    
# 단어별 
print(wordfreq)
print(doccnt1)

print(doccnt2)
```
