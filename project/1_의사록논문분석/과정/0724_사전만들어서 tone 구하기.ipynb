```python
up_dict = polar[polar['nbc'] > 1.3]
down_dict = polar[polar['nbc'] < 0.73]
```

```python
up = up_dict['n-gram']
up_list = list(up)
len(up_list)
```

```python
down = down_dict['n-gram']

down_list = list(down)
len(down_list)
```

```python
ngrams['tone'] = 0
count_p = 1
count_n = 1
for i in range(len(ngrams)):
    ngrams_li = ngrams.loc[i, 'n-gram'].split(',')
    for ngram in ngrams_li:
        if ngram in up_list:
            print('up :', ngram)
            count_p += 1
        if ngram in down_list:
            print('down : ', ngram)
            count_n += 1
        
    ngrams.loc[i, 'tone'] = (count_p / len(up_list))/(count_n/len(down_list))
    
    if i%20 == 0:
        print(i)
```

```python
date = ngrams['date']
date_list = list(set(date))
```

```python
tone_list=[]
for date in date_list:
    tmp=ngrams[ngrams['date'] == date]
    cnt_p=len(tmp[tmp['tone']>0])
    cnt_n=len(tmp[tmp['tone']<0])
    print(cnt_p,cnt_n)
    if (cnt_p + cnt_n) == 0:
        continue
    tone = (cnt_p-cnt_n) / (cnt_p+cnt_n)
    tone_list.append(tone)
    
    
```
