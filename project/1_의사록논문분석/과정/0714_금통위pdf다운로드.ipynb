```python

import requests
from bs4 import BeautifulSoup
base_url = 'http://www.bok.or.kr'

for i in range(1, 45):
    b_url = 'http://www.bok.or.kr/portal/bbs/B0000245/list.do?menuNo=200761'
    params = {
    'pageIndex' : i
    }
    
    resp = requests.get(b_url, params = params)
    soup = BeautifulSoup(resp.content)
    
    li_tags= soup.find('div', class_='bdLine type2').find('ul').find_all('li')
#     li_li = li.find_next_siblings('li')
#     li_list.append(li)
#     li_list.extend(li_li)
    li_list = []
    li_list.append(li_tags[0])
    for i in range(0,27,3):
        li_list.append(li_tags[i + 3])
        
    for li in li_list:
        print(li)
        url_to_li = li.find('a').attrs['href']
        follow_link_url = base_url + url_to_li
        link_resp = requests.get(follow_link_url)
        print(link_resp)
        print('-'*20)
        soup_link = BeautifulSoup(link_resp.content)
        
        title = soup_link.find('h3', class_ = 'subject').text
        
        li_all_link = soup_link.find('div', class_ = 'addfile').find('ul').find_all('li')
        
        for pdf_or_hw_li in li_all_link:
            if 'pdf' in pdf_or_hw_li.find('a').text:
                pdf_url = pdf_or_hw_li.find('a').attrs['href']
                dl_link = base_url + pdf_url
                print(dl_link)
                dl_resp = requests.get(dl_link)
                
                with open('{}.pdf'.format(title), 'wb') as f:
                    f.write(dl_resp.content)
                print('0' * 20)
            else:
                pass       
        url_to_li = ''
        follow_link_url = ''
        dl_link = ''
        pdf_url = ''
        title = ''
        count += 3
  
```
