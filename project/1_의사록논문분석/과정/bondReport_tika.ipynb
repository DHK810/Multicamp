```python
import re
from tika import parser #https://github.com/chrismattmann/tika-python
import shutil
from os import listdir
from os.path import isfile, join
```

```python
# 폴더내 pdf파일을 txt 파일로 변경
def pdf2txt (source_folder="./rawReport/", output_folder="./tikaReport/") :
    # 지정 폴더 내 파일 목록 조회 (파일만)
    pdf_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    
    for pdf in pdf_files :
        pdf_filepath = source_folder + pdf
        pdf_tmp_filepath = output_folder + 'tmp.pdf'
        # pdf 파일을 text로 변환

        
        shutil.copyfile(pdf_filepath, pdf_tmp_filepath)
        try:    
            parsedPDF = re.sub('\n', '', parser.from_file(pdf_tmp_filepath)["content"])
#         print(parsedPDF)
        except Exception as e:
            parasedPDF = ' '
            print(f'{pdf} 에러 : {e}')

        ouput_filepath = (output_folder + pdf).replace('.pdf', '.txt')
        with open(ouput_filepath, 'w', encoding='utf-8') as f :
            print(ouput_filepath)
            f.write(parsedPDF)
            f.close()
```

```python
pdf2txt()
```

```python
import shutil

def txtRename (source_folder="./tikaReport/", output_folder="./textRename/") :
    # 지정 폴더 내 파일 목록 조회 (파일만)
    txt_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    
    
    for txt in txt_files :
        txt_filepath = source_folder + txt
        txt_tmp_filepath = output_folder + 'tmp.txt'
        
        txt_list=[]
        tmp_txt=''
        tmp_txt=txt
        
        txt_list=tmp_txt.replace('.txt','').split('_')
        
        #앞부분 숫자 수정
        if int(txt_list[0]) < 10:
            txt_list[0]='00'+txt_list[0]
        elif int(txt_list[0]) < 100:
            txt_list[0]='0'+txt_list[0]
            
        #뒷부분 숫자 수정
        if int(txt_list[1]) < 10:
            txt_list[1]='0'+txt_list[1]
        
        new_txt='_'.join(txt_list)
        new_txt+='.txt'
        
        output_filepath = (output_folder + new_txt)
        
        with open(output_filepath, 'w', encoding='utf-8') as f :
            print(output_filepath)
            f.close()
            
        shutil.copyfile(txt_filepath, output_filepath)
```

```python
txtRename()
```
